Q13.How is memory managed in Python?
Memory in Python is managed by the Python memory manager, which involves:
Automatic garbage collection.
Reference counting.
Dynamic allocation of memory for variables.
Using private heaps for managing objects and data structures.

Q14.How to install Python on Windows and set path variables?
To install Python on Windows and set path variables:
Download Python Installer: Go to the official Python website and download the latest version of Python.
Run Installer: Run the downloaded installer.
Check "Add Python to PATH": Ensure that the "Add Python to PATH" checkbox is selected.
Install: Click "Install Now" and follow the prompts.
Verify Installation: Open Command Prompt and type python --version to verify the installation.
